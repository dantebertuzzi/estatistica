# Análise **Exploratória de Dados** (EDA)
Em estatística, a análise exploratória de dados (AED) é uma abordagem à análise de conjuntos de dados de modo a **resumir suas características principais**, frequentemente com **métodos visuais**. Um modelo estatístico pode ou não ser usado, mas primariamente a AED tem como objetivo observar o que os dados podem nos dizer além da modelagem formal ou do processo de teste de hipóteses.

A análise exploratória de dados foi promovida pelo estatístico norte-americano **John Tukey**, que incentivava os estatísticos a explorar os dados e possivelmente formular hipóteses que poderiam levar a novas coletas de dados e experimentos. A AED é diferente da análise inicial de dados (AID), que se concentra mais estreitamente em verificar os pressupostos exigidos para ajuste de modelos e teste de hipóteses, além de manusear `valores faltantes` e fazer transformações de variáveis conforme necessário. A análise exploratória de dados abrange a AID.

**John Tukey** definiu a análise de dados em **1961** como: *"Procedimentos para analisar dados, técnicas para interpretar os resultados de tais procedimentos, formas de planejar a reunião dos dados para tornar sua análise mais fácil, mais precisa ou mais exata e toda a maquinaria e os resultados da estatística (matemática) que se aplicam a análise de dados."*.

**A defesa da AED por Tukey** encorajou o desenvolvimento de pacotes de estatística computacional, especificamente a `S` na Bell Labs. A linguagem de programação `S` inspirou os sistemas `S-PLUS` e `R`. Esta família de ambientes de estatística computacional incluiu capacidades de *visualização dinâmica altamente melhoradas*, que permitiram aos estatísticos identificar valores atípicos, tendências e padrões em dados que mereciam estudos posteriores.

**A análise exploratória de dados de Tukey** estava relacionada com outros dois desenvolvimentos na teoria estatística: a *estatística robusta* e a *estatística não paramétrica*, ambas as quais tentavam reduzir a sensibilidade das inferências estatísticas a erros na formulação de modelos estatísticos. Tukey promoveu o uso do *resumo dos cinco números de dados* — os dois extremos (máximo e mínimo), a mediana e os quartis — já que a mediana e os quartis, sendo funções da *distribuição empírica*, são definidos para todas as distribuições, diferentemente da média e do desvio padrão. Além disso, a mediana e os quartis são mais *robustos* em relação a distribuições oblíquas e de cauda pesada do que os resumos tradicionais (a média e o desvio padrão). Os pacotes `S`, `S-PLUS` e `R` incluem rotinas usando *estatísticas de reamostragem*, tais como o `jackknife` de Quenouille e Tukey e o `bootstrapping` de Efron, que são não paramétricas e robustas (para muitos problemas).

**A AED**, a *estatística robusta*, a *estatística não paramétrica* e o desenvolvimento das *linguagens de programação estatística* facilitaram o trabalho dos estatísticos em problemas de ciência e engenharia. Tais problemas incluíam a fabricação de semicondutores e a compreensão de redes de comunicação, que interessavam à Bell Labs. Estes desenvolvimentos estatísticos, todos defendidos por Tukey, foram desenhados para complementar a teoria analítica dos *testes de hipóteses estatísticas*, particularmente a *tradição laplaciana de ênfase em famílias exponenciais*.

### **Desenvolvimento**


John W. Tukey escreveu o livro *Análise Exploratória de Dados* em 1977. Tukey sustentou que muita ênfase na estatística foi colocada no *teste de hipóteses estatísticas* (`análise de dados confirmatórios`); mais ênfase precisava ser colocada no uso de dados para sugerir hipóteses a serem testadas. Em particular, ele sustentou que *confundir os dois tipos de análises e empregá-los no mesmo conjunto de dados* pode levar a um *viés sistemático* devido aos problemas inerentes ao teste de hipóteses sugeridas pelos dados.

**Os objetivos da EDA são:**

- Habilitar descobertas inesperadas nos dados
- Sugerir hipóteses sobre as causas dos fenômenos observados
- Avaliar as suposições nas quais a inferência estatística será baseada
- Apoiar a seleção de ferramentas e técnicas estatísticas apropriadas
- Fornecer uma base para a recolha de dados adicionais através de inquéritos ou experiências

Muitas técnicas de `EDA` foram adotadas na *mineração de dados*. Elas também estão sendo ensinadas a jovens estudantes como uma forma de apresentá-los ao *pensamento estatístico*.


### Dados = Modelo + Resíduos: Uma Explicação segundo Tukey

A expressão **dados = modelo + resíduos** é uma representação fundamental em análise de dados, particularmente na análise de regressão, e pode ser associada à abordagem de John Tukey, um renomado estatístico que enfatizou a importância da visualização de dados e da análise exploratória.

$$
\text{Dados} = \text{Modelo} + \text{Resíduos}
$$

#### Explicação da Equação

1. **Dados**:
   - Refere-se às observações reais que coletamos. No contexto de um experimento ou de um estudo, os dados são os valores que medimos ou registramos. Por exemplo, se estivermos analisando a relação entre consumo e rendimento, os dados representariam as medições reais de consumo para cada nível de rendimento.

2. **Modelo**:
   - O modelo é uma função ou fórmula que descreve a relação entre as variáveis que estamos analisando. É uma tentativa de capturar o padrão subjacente nos dados. Por exemplo, um modelo linear pode ser representado como $y = mx + b$, onde $y$ é o consumo, $x$ é o rendimento, $m$ é a inclinação da linha de regressão, e $b$ é a interseção no eixo $y$. O modelo tenta prever ou estimar os valores de $y$ (consumo) com base nos valores de $x$ (rendimento).

3. **Resíduos**:
   - Os resíduos representam a diferença entre os dados observados e os valores previstos pelo modelo. Matematicamente, eles são definidos como:
     $$
     \text{Resíduo} = \text{Dados Observados} - \text{Modelo Previsto}
     $$
   - Os resíduos nos dizem o quanto o modelo falha em prever os dados reais. Um resíduo positivo indica que o modelo subestimou o valor observado, enquanto um resíduo negativo indica que o modelo superestimou.

#### Interpretação de Tukey

John Tukey, em seu trabalho sobre análise de dados, destacou a importância de entender a variabilidade nos dados e a relação entre os dados, o modelo e os resíduos. Ele enfatizou que:

- **Visualização**: A visualização dos dados é essencial para identificar padrões, outliers e a qualidade do ajuste do modelo. Gráficos de dispersão, boxplots e histogramas são ferramentas que ajudam a entender como os dados se distribuem e como se relacionam com o modelo.
  
- **Exploração de Dados**: A análise exploratória de dados é fundamental para entender a estrutura dos dados antes de aplicar modelos estatísticos. Tukey defendia que os analistas deveriam explorar as propriedades dos dados, como a presença de tendências, sazonalidade ou variabilidade.

- **Importância dos Resíduos**: A análise dos resíduos é crucial para avaliar a qualidade do modelo. Resíduos grandes ou padrões não aleatórios nos resíduos podem indicar que o modelo não está capturando adequadamente a relação subjacente nos dados, sugerindo a necessidade de um modelo diferente ou a inclusão de variáveis adicionais.

#### Conclusão

A relação **dados = modelo + resíduos** é uma maneira de entender como os modelos estatísticos tentam capturar a realidade dos dados. A análise cuidadosa e a interpretação dos resíduos, como enfatizado por Tukey, são fundamentais para garantir que o modelo escolhido seja apropriado e útil para a análise e previsões futuras.


```julia
using Plots, GLM, DataFrames

# Dados fictícios de consumo e rendimento com maior dispersão
rendimento = 1:10
consumo = [1.8, 3.5, 2.0, 4.8, 3.1, 6.3, 4.5, 7.8, 5.9, 9.2]  # Aumentamos a dispersão dos dados

# Criando um DataFrame com os dados
df = DataFrame(rendimento = rendimento, consumo = consumo)

# Ajustando um modelo de regressão linear
modelo = lm(@formula(consumo ~ rendimento), df)

# Prevendo os valores ajustados pelo modelo
consumo_previsto = predict(modelo)

# Calculando os resíduos (diferença entre dados e modelo)
residuos = df.consumo .- consumo_previsto

# Criando o plot com os dados reais
plot(df.rendimento, df.consumo, seriestype=:scatter, label="Dados Reais", legend=:topleft)
plot!(df.rendimento, consumo_previsto, label="Modelo Ajustado (Regressão Linear)", lw=2, color=:red)

# Adicionando as linhas verticais para cada ponto
for i in 1:length(df.rendimento)
    # Linha vertical indicando o dado real
    plot!([df.rendimento[i], df.rendimento[i]], [0, df.consumo[i]], label=false, color=:blue, linestyle=:dash)

    # Linha vertical indicando o modelo ajustado
    plot!([df.rendimento[i], df.rendimento[i]], [0, consumo_previsto[i]], label=false, color=:green, linestyle=:dot)

    # Linha conectando o dado ao modelo (resíduo) com menor espessura
    plot!([df.rendimento[i], df.rendimento[i]], [df.consumo[i], consumo_previsto[i]], label=false, color=:gray, linestyle=:solid, lw=1)
end

# Configuração de títulos e legendas
xlabel!("Rendimento")
ylabel!("Consumo")
title!("Relação entre Consumo e Rendimento\nDados = Modelo + Resíduos")

# Adicionando legenda para as linhas verticais
hline!([], label="Dados Reais", color=:blue, linestyle=:dash)
hline!([], label="Modelo Ajustado", color=:green, linestyle=:dot)
hline!([], label="Resíduos", color=:gray, linestyle=:solid)
```

### Tipos de Variáveis

Na estatística e na pesquisa, as variáveis podem ser classificadas de diferentes maneiras. As principais categorias de variáveis são:

#### 1. Variáveis **Qualitativas** (ou Categóricas)

**Variáveis qualitativas** representam categorias ou grupos e não têm um valor numérico intrínseco. Elas podem ser subdivididas em:

- **Nominais**: Não têm uma ordem específica. Exemplos incluem `sexo`, `cor do cabelo`, e `tipo de carro`.
  
- **Ordinais**: Possuem uma **ordem** ou **classificação**, mas a distância entre os valores não é mensurável. Exemplos incluem `nível de satisfação` (satisfeito, neutro, insatisfeito) e `faixa etária` (jovem, adulto, idoso).

#### 2. Variáveis **Quantitativas** (ou Numéricas)

**Variáveis quantitativas** representam valores numéricos e podem ser medidas ou contadas. Elas são divididas em:

- **Discretas**: Podem assumir apenas valores inteiros ou contáveis. Exemplos incluem `número de filhos`, `número de alunos em uma sala`, e `número de erros em um teste`.
  
- **Contínuas**: Podem assumir qualquer valor dentro de um intervalo, incluindo frações e decimais. Exemplos incluem `altura`, `peso`, e `temperatura`.

![IMAGEM](https://raw.githubusercontent.com/dantebertuzzi/public/refs/heads/main/Diagrama%20sem%20nome.drawio.png)

*Classificação de uma variável*

#### 3. Variáveis Independentes e Dependentes

- **Variáveis independentes**: Também conhecidas como variáveis explicativas, são aquelas que são manipuladas ou categorizadas para observar seu efeito em outra variável. Exemplos incluem `tratamento em um experimento` ou `tempo de exposição`.

- **Variáveis dependentes**: São as variáveis que são medidas ou observadas em um experimento, e que se espera que sejam afetadas pelas variáveis independentes. Exemplos incluem `resultado do teste` ou `nível de estresse`.

#### 4. Variáveis Controles

**Variáveis de controle** são aquelas que são mantidas constantes durante um experimento para evitar que influenciem os resultados. Elas ajudam a garantir que a relação entre a variável independente e a dependente seja clara.

#### Conclusão

Compreender os diferentes tipos de variáveis é essencial para a condução de análises estatísticas e para a interpretação adequada dos dados. Cada tipo de variável requer técnicas específicas para análise e interpretação, contribuindo para uma melhor compreensão dos fenômenos estudados.

## Distribuição de Frequências em Estatística

> Quando se estuda uma variável, o maior interesse do pesquisador é conhecer a distribuição dessa variável através das possíveis realizações (valores) da mesma.

A distribuição de frequências é uma tabela que mostra como os dados estão distribuídos ao longo de diferentes valores ou intervalos. Ela é utilizada para organizar e resumir grandes conjuntos de dados, permitindo uma análise mais fácil e visual.

### Principais Conceitos

# $n_i$

1. **Frequência Absoluta ($n_i$)**:
- Representa o número de ocorrências de um valor ou intervalo específico nos dados.
- Exemplo: Se 8 funcionários possuem Ensino Fundamental, a frequência absoluta é 8.

# $f_i$

2. **Proporção ($f_i$)**:
- É a relação entre a frequência absoluta e o total de observações, dada pela fórmula:
$$f_i = \frac{n_i}{N}$$
onde $N$ é o total de observações.
- Exemplo: Se 8 funcionários possuem Ensino Fundamental em um total de 36 funcionários, a proporção é:
$$f_i = \frac{8}{36} \approx 0.22$$

# $100 \cdot f_i$

3. **Porcentagem ($100 \cdot f_i$)**:
- É a proporção multiplicada por 100, expressa como uma porcentagem.
- Exemplo: Uma proporção de 0.22 equivale a 22%.

## Exemplo de Distribuição de Frequências

A tabela a seguir mostra a distribuição de frequências do grau de instrução de 36 empregados da seção de orçamentos da companhia Milsa:

| Grau de Instrução       | Contagem | Frequência Absoluta $n_i$ | Proporção $f_i$ | Porcentagem $100 \cdot f_i$ |
|-------------------------|----------|--------------------------|----------------|----------------------------|
| Ensino Fundamental      | 8        | $\frac{8}{36}$           | 0.22           | 22%                        |
| Ensino Médio            | 15       | $\frac{15}{36}$          | 0.42           | 42%                        |
| Ensino Superior         | 10       | $\frac{10}{36}$          | 0.28           | 28%                        |
| Pós-graduação           | 3        | $\frac{3}{36}$           | 0.08           | 8%                         |
| **Total**               | **36**   | **$\frac{36}{36} = 1$**  | **1.00**       | **100%**                   |

### Interpretação da Tabela

- **Ensino Fundamental**: 8 funcionários, o que representa `22%` do total.
- **Ensino Médio**: 15 funcionários, correspondendo a `42%` do total.
- **Ensino Superior**: 10 funcionários, equivalendo a `28%` do total.
- **Pós-graduação**: 3 funcionários, representando `8%` do total.

### Como Calcular Frequências e Proporções

1. **Frequência Absoluta ($n_i$)**:
   - Conte o número de ocorrências de cada categoria.

2. **Proporção ($f_i$)**:
   - Use a fórmula:
   $$f_i = \frac{n_i}{N}$$
   onde $n_i$ é a frequência da categoria e $N$ é o total de observações.

3. **Porcentagem**:
   - Multiplique a proporção por 100:
   $$100 \cdot f_i$$

### Importância da Distribuição de Frequências

- **Organização dos Dados**: Simplifica a visualização de como os dados estão distribuídos.
- **Identificação de Padrões**: Ajuda a identificar tendências e padrões nos dados.
- **Base para Análise Estatística**: Facilita a execução de análises mais complexas.

```julia
using DataFrames

# Criando o DataFrame
df = DataFrame(
    Nivel_Educacional = ["Ensino Fundamental", "Ensino Médio", "Ensino Superior", "Pós-graduação"],
    Frequencia_Absoluta = [8, 15, 10, 3]
)

# Calculando a frequência relativa e a porcentagem
df.Frequencia_Relativa = df.Frequencia_Absoluta ./ sum(df.Frequencia_Absoluta)
df.Porcentagem = df.Frequencia_Relativa .* 100

# Exibindo o DataFrame com os resultados
println(df)
```

```julia
# Calculando a frequência acumulada absoluta e relativa
df.Frequencia_Acumulada_Absoluta = cumsum(df.Frequencia_Absoluta)
df.Frequencia_Acumulada_Relativa = cumsum(df.Frequencia_Relativa)

# Exibindo o DataFrame final com todas as colunas de análise
println(df)
```

```julia
using Plots

# Gráfico de barras para Frequência Absoluta
bar(df.Nivel_Educacional, df.Frequencia_Absoluta, 
    title = "Frequência Absoluta por Nível Educacional",
    xlabel = "Nível Educacional",
    ylabel = "Frequência Absoluta",
    legend = false,
    color = :blue)

# Gráfico de linhas para Frequência Acumulada Relativa
plot(df.Nivel_Educacional, df.Frequencia_Acumulada_Relativa, 
     seriestype = :line,
     title = "Frequência Acumulada Relativa por Nível Educacional",
     xlabel = "Nível Educacional",
     ylabel = "Frequência Acumulada Relativa",
     marker = :circle,
     color = :red,
     legend = false)
```

```julia
using Statistics

# Calculando estatísticas descritivas
media = mean(df.Frequencia_Absoluta)
mediana = median(df.Frequencia_Absoluta)
variancia = var(df.Frequencia_Absoluta)
desvio_padrao = std(df.Frequencia_Absoluta)

# Exibindo os resultados
println("Média da Frequência Absoluta: ", media)
println("Mediana da Frequência Absoluta: ", mediana)
println("Variância da Frequência Absoluta: ", variancia)
println("Desvio Padrão da Frequência Absoluta: ", desvio_padrao)
```
