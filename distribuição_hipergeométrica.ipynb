1. Em um laboratório de análises clínicas foram realizados `54` exames de HIV, dos quais `14 foram soropositivos`. Um inspetor irá pegar `5` destes exames de uma maneira casual. Qual será a probabilidade de número de exames soropositivos nesta amostra seja:

a) **– 2**

b) **– 5**

>Esse problema pode ser resolvido utilizando a `distribuição hipergeométrica`, que é apropriada quando temos uma `população finita` e não estamos repondo os elementos após a amostragem `(sem reposição)`. A distribuição hipergeométrica calcula a probabilidade de um número específico de sucessos (neste caso, exames soropositivos) em uma amostra aleatória, dado o número total de sucessos na população.

>###### Resposta

#### 1. Informações do Problema
- **Total de exames**: $N = 54$
- **Exames soropositivos**: $K = 14$
- **Tamanho da amostra**: $n = 5$
- Precisamos calcular as probabilidades para:
  - $k = 2$ soropositivos na amostra (opção A)
  - $k = 5$ soropositivos na amostra (opção B)

#### 2. Fórmula da Distribuição Hipergeométrica

A fórmula da distribuição hipergeométrica é:

$$
P(X = k) = \frac{\binom{K}{k} \binom{N - K}{n - k}}{\binom{N}{n}}
$$

Onde:
- $\binom{K}{k}$ é o número de combinações de $K$ soropositivos selecionando $k$,
- $\binom{N - K}{n - k}$ é o número de combinações de $N - K$ soronegativos selecionando $n - k$,
- $\binom{N}{n}$ é o número total de maneiras de escolher $n$ exames entre os $N$.

#### 3. Cálculo da Probabilidade para $k = 2$

Para $k = 2$ soropositivos na amostra, aplicamos a fórmula:

$$
P(X = 2) = \frac{\binom{14}{2} \binom{40}{3}}{\binom{54}{5}}
$$

Calculando as combinações:
- $\binom{14}{2} = \frac{14 \times 13}{2!} = 91$
- $\binom{40}{3} = \frac{40 \times 39 \times 38}{3!} = 9,880$
- $\binom{54}{5} = \frac{54 \times 53 \times 52 \times 51 \times 50}{5!} = 3,162,510$

Substituindo os valores:

$$
P(X = 2) = \frac{91 \times 9,880}{3,162,510} = 0.2843
$$

Portanto, a probabilidade de haver exatamente 2 soropositivos na amostra é **28,43%**.

#### 4. Cálculo da Probabilidade para $k = 5$

Para $k = 5$ soropositivos na amostra, aplicamos a fórmula:

$$
P(X = 5) = \frac{\binom{14}{5} \binom{40}{0}}{\binom{54}{5}}
$$

Calculando as combinações:
- $\binom{14}{5} = \frac{14 \times 13 \times 12 \times 11 \times 10}{5!} = 2,002$
- $\binom{40}{0} = 1$ (pois estamos escolhendo 0 soronegativos)
- $\binom{54}{5} = 3,162,510$ (mesmo valor de antes)

Substituindo os valores:

$$
P(X = 5) = \frac{2,002 \times 1}{3,162,510} = 0.00063
$$

Portanto, a probabilidade de haver exatamente 5 soropositivos na amostra é **0,063%**.

#### 5. Conclusão

- **Probabilidade de $k = 2$** soropositivos: $0.2843$ ou **28,43%**.
- **Probabilidade de $k = 5$** soropositivos: $0.00063$ ou **0,063%**.
