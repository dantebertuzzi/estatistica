# Guia Conceitual: **Correlação**

## O que é Correlação?

**Correlação** é uma medida estatística que descreve a força e a direção de uma relação linear entre duas variáveis. Em termos simples, a correlação indica o grau em que duas variáveis se movem em conjunto — ou seja, como as mudanças em uma variável estão relacionadas às mudanças em outra.

### História e Origem da Correlação

#### Origem do Conceito de Correlação

O conceito de correlação foi introduzido no final do século XIX pelo cientista britânico **Sir Francis Galton**. Galton era um polímata, interessado em diversas áreas, como a estatística, biologia e a psicologia. Ele foi um dos pioneiros no estudo das relações entre variáveis e no desenvolvimento de métodos estatísticos para quantificar essas relações.

#### Francis Galton e a "Regressão à Média"

Galton é mais conhecido por seu trabalho sobre a **regressão à média**, um conceito relacionado à correlação. Ele observou que os filhos de pais altos tendiam a ser mais baixos que seus pais, e os filhos de pais baixos tendiam a ser mais altos — um retorno à média da altura da população. Esse fenômeno foi uma das primeiras aplicações práticas da ideia de correlação.

Galton também foi o primeiro a desenvolver uma forma rudimentar de **diagrama de dispersão** (scatter plot) para representar graficamente a relação entre duas variáveis.

#### Karl Pearson e o Coeficiente de Correlação

Embora Galton tenha feito descobertas iniciais sobre correlação, foi o matemático e estatístico **Karl Pearson** quem formalizou o conceito e desenvolveu o **coeficiente de correlação de Pearson**, que hoje é amplamente utilizado para medir a força e a direção de uma relação linear entre duas variáveis. Pearson era um colaborador de Galton e estendeu seus trabalhos, criando um sistema mais rigoroso e matemático para quantificar essas relações.

O coeficiente de correlação de Pearson, também conhecido como **"r" de Pearson**, foi proposto por ele no início do século XX e é até hoje uma das ferramentas estatísticas mais utilizadas.

## Coeficiente de Correlação (r)

O coeficiente de correlação, denotado por **r**, varia de -1 a 1 e indica:

- **r = 1**: Correlação positiva perfeita. À medida que uma variável aumenta, a outra também aumenta proporcionalmente.
- **r = -1**: Correlação negativa perfeita. À medida que uma variável aumenta, a outra diminui proporcionalmente.
- **r = 0**: Nenhuma correlação. Não há relação linear aparente entre as variáveis.

### Fórmula do Coeficiente de Correlação de Pearson
A fórmula mais comum para calcular a correlação é o **coeficiente de correlação de Pearson**, dado por:

$r = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2} \sum{(y_i - \bar{y})^2}}}$

Onde:
- $x_i$ e $y_i$ são os valores das variáveis.
- $\bar{x}$ e $\bar{y}$ são as médias das variáveis.

## Tipos de Correlação

1. **Correlação Positiva**: Quando o valor de **r** está próximo de 1, significa que as duas variáveis tendem a aumentar juntas. Exemplo: A altura e o peso de uma pessoa geralmente têm uma correlação positiva.

2. **Correlação Negativa**: Quando o valor de **r** está próximo de -1, significa que uma variável tende a diminuir à medida que a outra aumenta. Exemplo: O preço de um produto e a demanda podem ter uma correlação negativa.

3. **Correlação Nula (ou Zero)**: Quando o valor de **r** é próximo de 0, significa que não há relação linear aparente entre as variáveis. Isso não implica que não haja relação, apenas que ela não é linear.

## Interpretação dos Valores de r

- **0.7 ≤ |r| ≤ 1.0**: Forte correlação
- **0.3 ≤ |r| < 0.7**: Correlação moderada
- **0.0 ≤ |r| < 0.3**: Fraca correlação ou nenhuma

## Limitações da Correlação

1. **Somente Relações Lineares**: A correlação de Pearson mede apenas relações lineares. Se a relação entre as variáveis for não linear, o coeficiente de correlação pode ser baixo, mesmo que as variáveis estejam fortemente relacionadas de outras maneiras.

2. **Sensibilidade a Outliers**: Valores extremos (outliers) podem distorcer a correlação, dando uma impressão incorreta da relação entre as variáveis.

3. **Causalidade**: Correlação não implica causalidade. Mesmo que duas variáveis estejam correlacionadas, isso não significa que uma causa a outra.

## Correlação vs. Causalidade

Um ponto importante ao interpretar a correlação é lembrar que **correlação não implica causalidade**. Duas variáveis podem estar correlacionadas devido a uma causa comum, coincidência, ou outras variáveis não observadas (chamadas de variáveis de confusão). Um exemplo clássico de correlação sem causalidade é o aumento do consumo de sorvete no verão e o aumento de crimes violentos — ambos aumentam no verão, mas não porque um causa o outro.

## Métodos de Cálculo da Correlação

Além do coeficiente de correlação de Pearson, existem outros métodos de calcular a correlação, como:

- **Correlação de Spearman**: Mede a correlação com base nas classificações das variáveis, em vez de seus valores exatos. É mais apropriada para dados ordinais ou relações não lineares.
  
- **Correlação de Kendall**: Similar à correlação de Spearman, mas baseada na concordância entre pares de observações.

## Aplicações da Correlação

A correlação é amplamente usada em várias áreas, como:

- **Ciências Sociais**: Para estudar a relação entre variáveis como educação e renda.
- **Economia**: Para analisar a relação entre variáveis macroeconômicas, como inflação e desemprego.
- **Finanças**: Para examinar a relação entre o desempenho de diferentes ativos financeiros.
- **Medicina**: Para investigar a relação entre hábitos de vida e o risco de doenças.

## Conclusão

A correlação é uma ferramenta valiosa para entender as relações entre variáveis. Embora seja simples de calcular e interpretar, deve-se ter cuidado ao analisar os resultados, lembrando-se das limitações da correlação, especialmente no que diz respeito à causalidade.
