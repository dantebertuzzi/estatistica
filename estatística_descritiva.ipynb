# Guia Conceitual: Média **Aritmética**

## Origem do Conceito de Média Aritmética

O termo "aritmética" tem suas raízes no termo latino *arithmetica*, que deriva das palavras gregas `ἀριθμός (arithmos)`, que significa **número**, e `ἀριθμητική τέχνη (arithmetike tekhne)`, que significa **a arte de contar**.

A **média aritmética** é um dos conceitos mais antigos e amplamente utilizados em estatística e matemática. Sua origem remonta aos antigos babilônios e gregos, que usavam a média para calcular taxas e proporções em problemas cotidianos. Ao longo da história, a média se tornou uma ferramenta fundamental para resumir conjuntos de dados e entender tendências centrais.

Embora a média aritmética seja frequentemente usada para relatar tendências centrais, ela não é uma estatística robusta: **é muito influenciada por valores discrepantes** (valores muito maiores ou menores do que a maioria dos outros).

### Definição

> A média aritmética de um conjunto de dados observados é igual à soma dos valores numéricos de cada observação, dividida pelo número total de observações.

Em outras palavras, a média aritmética é o **valor central** de um conjunto de dados, calculado pela soma de todos os valores dividida pela quantidade de valores. Ela fornece uma medida simples da "tendência central" dos dados.

### Fórmula Geral
A fórmula para calcular a média aritmética é dada por:

$$
\text{Média Aritmética} = \frac{\sum_{i=1}^{n} x_i}{n}
$$

Onde:
- $x_i$ são os valores individuais do conjunto de dados.
- $n$ é o número total de valores.

## Diferença entre Média Amostral e Média Populacional

Existem dois tipos de médias que são comumente usadas: a **média amostral** e a **média populacional**. Cada uma delas tem uma notação e interpretação diferente, conforme descrito abaixo.

### Média Amostral ($\bar{x}$)
A **média amostral** é a média calculada a partir de uma **amostra** dos dados, que representa uma porção da população total.

- **Símbolo**: $\bar{x}$ (lê-se "x barra")
- **Fórmula**: 
$$
\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}
$$
Onde:
  - $\bar{x}$ é a média da amostra.
  - $x_i$ são os valores individuais da amostra.
  - $n$ é o número de observações na amostra.

### Média Populacional ($\mu$)
A **média populacional** é a média calculada a partir de **todos** os membros da população.

- **Símbolo**: $\mu$ (letra grega "mu")
- **Fórmula**: 
$$
\mu = \frac{\sum_{i=1}^{N} x_i}{N}
$$
Onde:
  - $\mu$ é a média da população.
  - $x_i$ são os valores individuais da população.
  - $N$ é o número total de observações na população.

## Diferença Fundamental
- A **média amostral** é usada quando você trabalha com uma amostra dos dados, que é apenas uma parte da população total.
- A **média populacional** é usada quando você tem dados de toda a população.

## Exemplo Prático

Suponha que você tenha os seguintes valores representando as notas de 5 alunos:

$$
\{7, 8, 6, 9, 10\}
$$

A média aritmética será calculada assim:

$$
\bar{x} = \frac{7 + 8 + 6 + 9 + 10}{5} = \frac{40}{5} = 8
$$

Neste exemplo, a média das notas é 8.

## Propriedades da Média Aritmética

1. **Soma dos Desvios**: A soma dos desvios (diferença de cada valor para a média) de todos os valores em relação à média aritmética é sempre igual a zero.
2. **Influência de Valores Extremos**: A média aritmética é sensível a valores muito altos ou muito baixos (outliers). Isso significa que um valor extremo pode distorcer a média.
3. **Média como Ponto de Equilíbrio**: A média pode ser vista como um ponto de equilíbrio dos dados. Se você imaginar os dados dispostos em uma linha, a média é o ponto onde todos os pesos dos dados se equilibram.

## Limitações da Média Aritmética

Embora seja amplamente utilizada, a média aritmética possui limitações:
- **Outliers**: Valores muito diferentes dos outros podem distorcer a média.
- **Distribuições Assimétricas**: Em distribuições com forte assimetria, a média pode não representar adequadamente o centro dos dados.

## Contraste com mediana
A média aritmética pode ser contrastada com a mediana. A mediana é definida de forma que não mais da metade dos valores sejam maiores e não mais da metade sejam menores do que ela. **Se os elementos nos dados aumentam aritmeticamente quando colocados em alguma ordem, a mediana e a média aritmética são iguais**. Por exemplo, considere a amostra de dados ${1,2,3,4}$. A média é ${2.5}$, assim como a mediana. No entanto, quando consideramos uma amostra que não pode ser organizada para aumentar aritmeticamente, como ${1,2,4,8,16}$, a mediana e a média aritmética podem diferir significativamente. Nesse caso, a média aritmética é $6.2$, enquanto a mediana é $4$. O valor médio pode variar consideravelmente da maioria dos valores da amostra e pode ser maior ou menor do que a maioria.

## Conclusão

A média aritmética é uma medida estatística central e útil, amplamente usada para resumir conjuntos de dados. No entanto, ao interpretá-la, é importante considerar o contexto dos dados e a possibilidade de distorções causadas por valores extremos ou distribuições assimétricas.


```julia
# Importar bibliotecas necessárias
using Plots
using Statistics

# Dados exemplo
dados = [7, 8, 6, 9, 10]

# Calcular a média aritmética
media = mean(dados)

# Mostrar a média
println("A média dos dados é: ", media)

# Criar um gráfico de dispersão com os dados
scatter(1:length(dados), dados, label = "Dados", xlabel = "Índice", ylabel = "Valores", 
        legend=:topright, color=:blue, markersize=8, title="Média Aritmética")

# Adicionar uma linha horizontal representando a média
hline!([media], label="Média", linestyle=:dash, color=:red, linewidth=2)

# Exibir o gráfico
display(current())
```

# Média Aritmética **Ponderada**

## Origem do Conceito de Média Ponderada

O conceito de **média ponderada** está enraizado na ideia de dar diferentes pesos ou relevâncias aos valores em um conjunto de dados. A palavra "ponderada" deriva do termo latino *ponderare*, que significa "pesar" ou "avaliar". A média ponderada surgiu como uma maneira de refletir a importância relativa de cada observação, permitindo que certas informações sejam mais influentes no cálculo final.

A média ponderada é amplamente utilizada em várias áreas, como economia, finanças, estatística, e ciências sociais, especialmente quando é necessário considerar fatores de importância variável em um conjunto de dados.

### Definição

> A **média ponderada** de um conjunto de dados é uma média em que cada valor é multiplicado por um fator de peso, e a soma desses produtos é dividida pela soma dos pesos.

Em outras palavras, a média ponderada leva em consideração a importância de cada valor ao calcular a tendência central de um conjunto de dados, oferecendo uma medida mais precisa quando as contribuições dos dados são desiguais.

### Fórmula Geral

A fórmula para calcular a média ponderada é dada por:

$$
\text{Média Ponderada} = \frac{\sum_{i=1}^{n} x_i \cdot w_i}{\sum_{i=1}^{n} w_i}
$$

Onde:
- $x_i$ são os valores individuais do conjunto de dados.
- $w_i$ são os pesos correspondentes a cada valor.
- $n$ é o número total de valores.

## Diferença entre Média Aritmética e Média Ponderada

Enquanto a **média aritmética** calcula o valor central dos dados assumindo que todos os valores têm a mesma importância, a **média ponderada** ajusta o cálculo para refletir a relevância de cada valor. Ou seja, a média ponderada é utilizada quando cada valor contribui de maneira diferente para o resultado final.

### Quando usar a Média Ponderada

A média ponderada é ideal para situações onde diferentes valores têm diferentes níveis de importância ou frequência. Exemplos comuns incluem:
- **Notas de provas** onde provas finais têm mais peso que testes intermediários.
- **Preços médios** em economia, onde quantidades diferentes de produtos são compradas.
- **Índices de ações**, onde cada ação tem um impacto diferente no índice global.

## Exemplo Prático

Suponha que um aluno tenha realizado três provas com os seguintes pesos e notas:
- Prova 1: Nota = 7, Peso = 2
- Prova 2: Nota = 9, Peso = 3
- Prova 3: Nota = 8, Peso = 5

A média ponderada é calculada assim:

$$
\text{Média Ponderada} = \frac{(7 \cdot 2) + (9 \cdot 3) + (8 \cdot 5)}{2 + 3 + 5}
$$
$$
\text{Média Ponderada} = \frac{14 + 27 + 40}{10} = \frac{81}{10} = 8,1
$$

Portanto, a média ponderada das notas é 8,1, levando em consideração a importância relativa de cada prova.

## Propriedades da Média Ponderada

1. **Influência dos Pesos**: Quanto maior o peso de um valor, maior será sua influência sobre a média ponderada.
2. **Equivalência à Média Aritmética**: Se todos os pesos forem iguais, a média ponderada é equivalente à média aritmética.
3. **Flexibilidade**: A média ponderada permite uma análise mais precisa quando se trabalha com dados onde certas observações são mais significativas.

## Limitações da Média Ponderada

Embora a média ponderada seja uma ferramenta poderosa, ela possui algumas limitações:
- **Escolha dos Pesos**: A escolha dos pesos pode ser subjetiva, e diferentes conjuntos de pesos podem levar a resultados diferentes.
- **Complexidade de Cálculo**: Pode ser mais complexa de calcular e interpretar, especialmente quando há muitos valores e pesos.

## Contraste com Outras Médias

A média ponderada é frequentemente comparada com outras medidas de tendência central, como a média aritmética, a mediana e a média geométrica:
- **Média Aritmética**: Assume que todos os valores têm a mesma importância, ao contrário da média ponderada.
- **Mediana**: Não é influenciada por valores extremos ou pesos, mas não leva em consideração a magnitude dos valores.
- **Média Geométrica**: Usada em contextos onde a taxa de crescimento ou proporções são mais relevantes, mas não considera pesos específicos para valores.

## Aplicações da Média Ponderada

A média ponderada é amplamente utilizada em situações práticas, como:
- **Avaliação de desempenho**: Em empresas, onde as avaliações anuais consideram diferentes metas com pesos distintos.
- **Cálculos financeiros**: Como a média ponderada dos preços das ações, onde o volume de cada transação é considerado.
- **Estatísticas sociais e demográficas**: Onde diferentes grupos populacionais têm influências variadas em medidas agregadas.

## Conclusão

A média ponderada é uma ferramenta estatística essencial que fornece uma maneira mais equilibrada de calcular a tendência central quando os valores possuem diferentes níveis de importância. Ao utilizar a média ponderada, é crucial escolher os pesos de forma criteriosa para garantir que o cálculo final represente com precisão o impacto relativo de cada valor no conjunto de dados.


```julia
using StatsBase

# Definindo as notas e os pesos
notas = [7, 9, 8, 6, 10]
pesos = [2, 3, 5, 1, 4]

# Calculando a média ponderada
media_ponderada = mean(notas, Weights(pesos))

# Criar um gráfico de dispersão com os dados
scatter(1:length(notas), notas, label="Dados", xlabel="Índice", ylabel="Valores", 
        legend=:topright, color=:blue, markersize=8, title="Média Ponderada")

# Adicionar uma linha horizontal representando a média ponderada
hline!([media_ponderada], label="Média Ponderada", linestyle=:dash, color=:red, linewidth=2)

# Exibir o gráfico
display(current())
```

# Média **Geométrica**

## Origem do Conceito de Média Geométrica

A média geométrica é um conceito matemático fundamental, com suas raízes na geometria. O termo "geométrica" vem da palavra grega `γεωμετρία` (geometria), que significa **medição da terra**. Os antigos matemáticos gregos, como Pitágoras e Euclides, usavam princípios geométricos para resolver problemas de proporção e escalas, o que levou ao desenvolvimento da média geométrica.

A **média geométrica** é amplamente utilizada em diversas áreas, como finanças, estatística e ciências naturais, especialmente em situações onde se deseja calcular a taxa de crescimento ou a razão de mudança em uma série de dados que se multiplicam entre si.

Diferente da média aritmética, a média geométrica é **menos influenciada por valores extremos**, sendo uma medida mais adequada para dados que apresentam variação exponencial.

### Definição

> A média geométrica de um conjunto de dados observados é igual à raiz \( n \)-ésima do produto dos valores numéricos de cada observação, onde \( n \) é o número de observações.

Em outras palavras, a média geométrica é calculada multiplicando todos os valores e, em seguida, extraindo a raiz que corresponde ao número total de valores.

### Fórmula Geral

A fórmula para calcular a média geométrica de um conjunto de \( n \) valores é dada por:

$$
\text{Média Geométrica} = \sqrt[n]{\prod_{i=1}^{n} x_i}
$$

ou de forma equivalente:

$$
\text{Média Geométrica} = (x_1 \times x_2 \times \cdots \times x_n)^{\frac{1}{n}}
$$

Onde:
- $x_i$ são os valores individuais do conjunto de dados.
- $n$ é o número total de valores.

## Aplicações da Média Geométrica

A média geométrica é frequentemente utilizada para calcular a **taxa de crescimento média** ao longo do tempo, como:
- Taxa de retorno de investimentos.
- Crescimento populacional.
- Crescimento econômico.

Ela é útil nesses contextos porque calcula a taxa de crescimento proporcional que gera o mesmo resultado final que os valores de crescimento observados.

## Propriedades da Média Geométrica

1. **Proporcionalidade**: A média geométrica é apropriada para dados que são expressos em taxas ou índices, pois reflete mudanças proporcionais nos valores.
2. **Menos Influência de Outliers**: Comparada à média aritmética, a média geométrica é menos influenciada por valores extremamente altos ou baixos.
3. **Multiplicatividade**: Se os valores dos dados forem multiplicados por uma constante, a média geométrica também será multiplicada por essa constante.

## Diferença entre Média Geométrica e Média Aritmética

Enquanto a **média aritmética** é usada para encontrar a soma média de um conjunto de números, a **média geométrica** é usada para encontrar a taxa média de mudança. Em geral:
- A média geométrica é sempre menor ou igual à média aritmética para o mesmo conjunto de dados.
- Em distribuições simétricas, onde os dados têm uma relação multiplicativa, a média geométrica é mais representativa do que a média aritmética.

## Exemplo Prático: Cálculo da Média Geométrica

Suponha que você tenha os seguintes valores que representam a taxa de crescimento anual de uma população ao longo de três anos:

$$
\{1.05, 1.07, 1.03\}
$$

Esses valores representam, respectivamente, os fatores de crescimento para cada ano. Para entender melhor, se a população era 100 no primeiro ano, ao final do primeiro ano, a população será:

$$
100 \times 1.05 = 105
$$

No segundo ano, aplicamos o crescimento de 7% (fator 1.07):

$$
105 \times 1.07 \approx 112.35
$$

E, finalmente, no terceiro ano, aplicamos o crescimento de 3% (fator 1.03):

$$
112.35 \times 1.03 \approx 115.75
$$

### Cálculo da Média Geométrica

A média geométrica é uma medida que se usa quando se deseja calcular a média de taxas de crescimento, pois considera o efeito composto dessas taxas. A fórmula da média geométrica para um conjunto de \(n\) números \(x_1, x_2, \ldots, x_n\) é:

$$
\text{Média Geométrica} = \left(x_1 \times x_2 \times \ldots \times x_n\right)^{\frac{1}{n}}
$$

Neste caso, temos \(n = 3\) (pois temos três anos) e os valores são \(x_1 = 1.05\), \(x_2 = 1.07\), e \(x_3 = 1.03\).

#### Passo 1: Multiplicação dos Valores

Primeiro, multiplicamos os fatores de crescimento:

$$
1.05 \times 1.07 \times 1.03
$$

Vamos calcular isso passo a passo:

1. **Multiplicando os dois primeiros valores**:
   $$
   1.05 \times 1.07 = 1.1235
   $$

2. **Agora multiplicamos o resultado pelo terceiro valor**:
   $$
   1.1235 \times 1.03 \approx 1.1570845
   $$

Então, temos:

$$
\text{Produto total} = 1.1570845
$$

#### Passo 2: Raiz do Produto

Agora, aplicamos a fórmula da média geométrica, que requer a raiz do produto. Precisamos calcular a raiz cúbica (já que temos 3 valores):

$$
\text{Média Geométrica} = (1.1570845)^{\frac{1}{3}}
$$

#### Passo 3: Cálculo da Raiz Cúbica

Para calcular a raiz cúbica de \(1.1570845\), podemos usar uma calculadora ou métodos de cálculo, mas também podemos fazer isso manualmente com uma aproximação. 

O valor aproximado da raiz cúbica é:

$$
(1.1570845)^{\frac{1}{3}} \approx 1.049
$$

#### Passo 4: Interpretação do Resultado

Portanto, a média geométrica das taxas de crescimento é:

$$
\text{Média Geométrica} \approx 1.049
$$

Isso significa que, em média, a população cresceu aproximadamente \(4.9\%\) ao ano. Para converter o fator de crescimento em porcentagem, fazemos:

$$
\text{Taxa de Crescimento em \%} = (1.049 - 1) \times 100\% \approx 4.9\%
$$

### Conclusão

Neste exemplo, calculamos a média geométrica das taxas de crescimento populacional ao longo de três anos, que resultou em uma taxa média de crescimento de aproximadamente \(4.9\%\). A média geométrica é uma ferramenta útil para medir o crescimento quando as taxas são compostas, pois leva em conta o efeito cumulativo das taxas de crescimento.


## Limitações da Média Geométrica

Embora seja amplamente utilizada, a média geométrica possui algumas limitações:
- **Valores Negativos e Zeros**: A média geométrica não pode ser calculada se os dados contiverem valores negativos ou zero.
- **Aplicabilidade Limitada**: Não é apropriada para conjuntos de dados que não são multiplicativos por natureza.

## Conclusão

A média geométrica é uma medida essencial em estatística e outras disciplinas, especialmente útil para calcular taxas de crescimento e mudanças proporcionais. Ela oferece uma alternativa robusta à média aritmética em situações onde os dados têm uma relação multiplicativa ou estão sujeitos a variações exponenciais.


```julia
# Definindo os valores dos dados
dados = [7, 9, 8, 6, 10]

# Calculando a média geométrica
media_geometrica = geomean(dados)

# Mostrar a média geométrica
println("A média geométrica dos dados é: ", media_geometrica)

# Criar um gráfico de dispersão com os dados
scatter(1:length(dados), dados, label="Dados", xlabel="Índice", ylabel="Valores", 
        legend=:topright, color=:green, markersize=8, title="Média Geométrica")

# Adicionar uma linha horizontal representando a média geométrica
hline!([media_geometrica], label="Média Geométrica", linestyle=:dash, color=:red, linewidth=2)

# Exibir o gráfico
display(current())
```

## Média **Harmônica**

A média harmônica, geralmente representada por $\bar {h}$,  é uma medida estatística que é usada para calcular a média de taxas, especialmente quando essas taxas são inversamente proporcionais. É particularmente útil em situações onde estamos lidando com razões e proporções, como velocidade ou densidade.

### Definição

A média harmônica de um conjunto de $n$ números $x_1, x_2, \ldots, x_n$ é dada pela fórmula:

$$
\text{Média Harmônica} = \frac{n}{\frac{1}{x_1} + \frac{1}{x_2} + \ldots + \frac{1}{x_n}}
$$

### Passo a Passo para Calcular a Média Harmônica

Vamos usar um exemplo prático para ilustrar o cálculo da média harmônica.

#### Exemplo

Suponha que você tenha as seguintes velocidades de um carro em três trechos de uma viagem (em $km/h$):

$$
\{60, 90, 120\}
$$

Essas velocidades representam a velocidade em diferentes trechos, e queremos encontrar a média harmônica para entender a velocidade média em termos de tempo gasto.

#### Passo 1: Contar os Números

Primeiro, conte o número de valores. Neste caso, temos $n = 3$ velocidades.

#### Passo 2: Calcular os Recíprocos

Calcule o recíproco de cada um dos valores:

$$
\frac{1}{60}, \quad \frac{1}{90}, \quad \frac{1}{120}
$$

Os valores são:

- $$\frac{1}{60} \approx 0.01667$$
- $$\frac{1}{90} \approx 0.01111$$
- $$\frac{1}{120} \approx 0.00833$$

#### Passo 3: Somar os Recíprocos

Agora, some todos os recíprocos calculados:

$$
\frac{1}{60} + \frac{1}{90} + \frac{1}{120}
$$

Calculando:

1. **Encontrando um denominador comum**:
   O denominador comum para 60, 90 e 120 é 180.

   - $$\frac{1}{60} = \frac{3}{180}$$
   - $$\frac{1}{90} = \frac{2}{180}$$
   - $$\frac{1}{120} = \frac{1.5}{180}$$

   Portanto,

   $$
   \frac{1}{60} + \frac{1}{90} + \frac{1}{120} = \frac{3 + 2 + 1.5}{180} = \frac{6.5}{180}
   $$

2. **Resultado**:
   $$\frac{6.5}{180} \approx 0.03611$$

#### Passo 4: Calcular a Média Harmônica

Agora, aplique a fórmula da média harmônica:

$$
\text{Média Harmônica} = \frac{n}{\text{Soma dos Recíprocos}} = \frac{3}{\frac{6.5}{180}} = \frac{3 \times 180}{6.5} \approx \frac{540}{6.5} \approx 83.08
$$

### Interpretação do Resultado

Portanto, a média harmônica das velocidades é aproximadamente$83.0$ km/h. Isso significa que, ao considerar as distâncias percorridas a essas velocidades, a velocidade média ponderada é$83.0$ km/h.

## Relações com outras médias

Uma construção geométrica das três médias de Pitágoras de dois números, $a$ e $b$. A Média Harmônica é denotada por $H$ na cor roxa. O $Q$ denota a quarta média, a média quadrática.

A média harmônica é uma das três médias de Pitágoras. Para todos os conjuntos de dados positivos que contêm, pelo menos um par de valores distintos, a média harmônica é sempre a mínima das três médias, enquanto que a média aritmética é sempre a maior das três e a média geométrica está sempre no meio. Se todos os valores de um conjunto de dados não vazio são iguais, as três médias são sempre iguais umas às outras. Por exemplo, as médias harmônica, geométrica e aritmética de $2, 2, 2$ são todas $2$.

A média harmônica é o caso especial $M^{-1}$ da média potência.

Uma vez que a média harmônica de uma lista de números tende para o mínimo dos elementos da lista, ela tende (em comparação com a média aritmética) a mitigar o impacto de grandes valores atípicos e agravar o impacto das pequenas.

A média aritmética é muitas vezes utilizada erroneamente em locais que exigem a média harmônica. Um exemplo é o cálculo da `velocidade média` em um percurso de ida e volta em uma mesma via, em que a ida é percorrida a 60 km/h e a volta a 40 km/h. A média aritmética de 50 está incorreta. A velocidade média no percurso total é a média harmônica de 40 e 60, ou seja 48 km/h. Isto se deve ao fato de que, como os dois trechos têm o mesmo comprimento, quanto menor for a velocidade, mais do tempo total é despendido àquela velocidade e, então, ela tem um peso maior na composição da velocidade média.

Se um conjunto de números não-idênticos é submetido a um `espalhamento` que preserve a média, ou seja, dois ou mais elementos do conjunto são `afastados` do outro, deixando a média aritmética inalterada, a média harmônica sempre diminui.


### Conclusão

A média harmônica é especialmente útil quando se trabalha com taxas e proporções. Ao usar a média harmônica, você obtém uma visão mais precisa da média quando as quantidades estão em relação inversa, como no caso de velocidades ou densidades. Essa média é menor que a média aritmética, refletindo a influência das taxas mais baixas no cálculo.


```julia
# Passo 1: Definir os valores dos dados
velocidades = [60, 90, 120]

# Passo 2: Calcular a média harmônica
n = length(velocidades)  # número de observações
soma_inversas = sum(1 ./ velocidades)  # soma das inversas
media_harmonica = n / soma_inversas  # cálculo da média harmônica
println("A média harmônica das velocidades é: ", media_harmonica)

# Passo 3: Criar um gráfico de dispersão com os dados
scatter(1:length(velocidades), velocidades, label="Velocidades", xlabel="Trecho", ylabel="Velocidade (km/h)", 
        legend=:topright, color=:blue, markersize=8, title="Média Harmônica das Velocidades")

# Passo 4: Adicionar uma linha horizontal representando a média harmônica
hline!([media_harmonica], label="Média Harmônica", linestyle=:dash, color=:red, linewidth=2)

# Exibir o gráfico
display(current())
```

### Média Harmônica Ponderada

A média harmônica ponderada é uma extensão da média harmônica, onde cada valor tem um peso associado. Essa média é útil em situações onde queremos levar em conta a importância relativa de diferentes observações.

### Definição

A média harmônica ponderada de um conjunto de $n$ valores $x_1, x_2, \ldots, x_n$ com pesos $w_1, w_2, \ldots, w_n$ é dada pela fórmula:

$$
H_w = \frac{\sum_{i=1}^{n} w_i}{\sum_{i=1}^{n} \frac{w_i}{x_i}}
$$

Onde:
- $H_w$ é a média harmônica ponderada.
- $w_i$ é o peso associado ao valor $x_i$.

### Propriedades

- A média harmônica ponderada é sempre menor ou igual à média aritmética ponderada.
- A média harmônica ponderada é mais sensível a valores menores, o que a torna útil em contextos onde as pequenas quantidades são mais significativas.

## Exemplo Prático: Média Harmônica Ponderada

### Contexto

Suponha que você tenha as seguintes notas de um aluno e seus respectivos pesos:

- **Notas:** \([7, 9, 8, 6, 10]\)
- **Pesos:** \([2, 3, 5, 1, 4]\)

### Cálculo da Média Harmônica Ponderada

A média harmônica ponderada é calculada utilizando a fórmula:

$$
H_w = \frac{\sum_{i=1}^{n} w_i}{\sum_{i=1}^{n} \frac{w_i}{x_i}}
$$

onde:
- $H_w$ é a média harmônica ponderada.
- $w_i$ é o peso associado à nota $x_i$.

### Passo a Passo do Cálculo

1. **Calcular os Pesos Totais:**

   $$
   \text{Total de Pesos} = 2 + 3 + 5 + 1 + 4 = 15
   $$

2. **Calcular a Soma dos Pesos Divididos pelas Notas:**

   $$
   \text{Soma} = \frac{2}{7} + \frac{3}{9} + \frac{5}{8} + \frac{1}{6} + \frac{4}{10}
   $$

   - $\frac{2}{7} \approx 0.2857$
   - $\frac{3}{9} = 0.3333$
   - $\frac{5}{8} = 0.6250$
   - $\frac{1}{6} \approx 0.1667$
   - $\frac{4}{10} = 0.4000$

   $$
   \text{Soma} \approx 0.2857 + 0.3333 + 0.6250 + 0.1667 + 0.4000 \approx 1.8107
   $$

3. **Calcular a Média Harmônica Ponderada:**

   $$
   H_w = \frac{15}{1.8107} \approx 8.27
   $$

### Interpretação

A média harmônica ponderada das notas é aproximadamente $8.27$. Isso significa que, considerando os pesos das notas, a média reflete melhor a importância de cada nota no conjunto.

### Conclusão

A média harmônica ponderada é uma ferramenta útil quando se deseja calcular uma média que leve em conta a significância de diferentes observações, especialmente em contextos onde as quantidades são variáveis.


```julia
# Definindo as notas e os pesos
notas = [7, 9, 8, 6, 10]
pesos = [2, 3, 5, 1, 4]

# Função para calcular a média harmônica ponderada
function media_harmonica_ponderada(notas, pesos)
    # Verificar se as notas e os pesos têm o mesmo tamanho
    if length(notas) != length(pesos)
        throw(ArgumentError("Notas e pesos devem ter o mesmo tamanho."))
    end

    # Calcular a soma dos pesos
    total_pesos = sum(pesos)
    
    # Calcular a soma dos pesos divididos pelas notas
    soma = sum(p / x for (p, x) in zip(pesos, notas))
    
    # Calcular a média harmônica ponderada
    return total_pesos / soma
end

# Calcular a média harmônica ponderada
media_ponderada = media_harmonica_ponderada(notas, pesos)

# Mostrar a média ponderada
println("A média harmônica ponderada das notas é: ", media_ponderada)

# Criar um gráfico de dispersão com os dados
scatter(1:length(notas), notas, label="Notas", xlabel="Índice", ylabel="Valores", 
        legend=:topright, color=:blue, markersize=8, title="Média Harmônica Ponderada")

# Adicionar uma linha horizontal representando a média harmônica ponderada
hline!([media_ponderada], label="Média Harmônica Ponderada", linestyle=:dash, color=:red, linewidth=2)

# Exibir o gráfico
display(current())
```

> ### Breve comparação das medidas de tendências centrais

```julia
# Gerar dados aleatórios
data = randn(1000)  # 1000 valores aleatórios da distribuição normal
weights = rand(1000)  # Pesos aleatórios para cada valor

# Calcular a mediana, médias e moda
mediana = median(data)
media_aritmetica = mean(data)
media_geometrica = exp(mean(log.(data[data .> 0])))  # Média geométrica apenas para dados positivos
media_harmonica = length(data) / sum(1.0 ./ data)     # Média harmônica
media_ponderada = sum(data .* weights) / sum(weights)  # Média ponderada
moda = mode(data)

# Criar o histograma
histogram(data, bins=30, alpha=0.6, legend=false,
          title="Histograma com Mediana, Médias e Moda",
          xlabel="Valores",
          ylabel="Frequência",
          color=:blue)

# Adicionar linhas da mediana, médias e moda
vline!([mediana], color=:red, label="Mediana", linewidth=2)
vline!([media_aritmetica], color=:green, label="Média Aritmética", linewidth=2)
vline!([media_geometrica], color=:orange, label="Média Geométrica", linewidth=2)
vline!([media_harmonica], color=:purple, label="Média Harmônica", linewidth=2)
vline!([media_ponderada], color=:magenta, label="Média Ponderada", linewidth=2)
vline!([moda], color=:cyan, label="Moda", linewidth=2)

# Exibir as anotações no gráfico em alturas diferentes
annotate!([(mediana, 50, text("Mediana: $mediana", :red)),
            (media_aritmetica, 60, text("Média Aritmética: $media_aritmetica", :green)),
            (media_geometrica, 40, text("Média Geométrica: $media_geometrica", :orange)),
            (media_harmonica, 70, text("Média Harmônica: $media_harmonica", :purple)),
            (media_ponderada, 80, text("Média Ponderada: $media_ponderada", :magenta)),
            (moda, 30, text("Moda: $moda", :cyan))])
```
