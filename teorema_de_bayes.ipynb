O **Teorema de Bayes** é um princípio fundamental da estatística que descreve a maneira como atualizamos nossas crenças em relação a um evento, **com base em novas evidências**. Ele é amplamente utilizado em probabilidade e estatística, principalmente em problemas onde precisamos revisar previsões ou hipóteses à medida que recebemos mais informações.
## Fórmula do **Teorema de Bayes**

O Teorema de Bayes pode ser expresso matematicamente da seguinte forma:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Onde:
- $P(A|B)$ é a <span style="color:orange;">**probabilidade posterior**</span>: a probabilidade de $A$ acontecer dado que $B$ já aconteceu.
- $P(B|A)$ é a <span style="color:orange;">**probabilidade verossímil**</span>: a probabilidade de $B$ acontecer dado que $A$ é verdadeiro.
- $P(A)$ é a <span style="color:orange;">**probabilidade anterior**</span> ou **priori**: a probabilidade de $A$ acontecer sem qualquer informação adicional.
- $P(B)$ é a <span style="color:orange;">**probabilidade marginal**</span>: a probabilidade de $B$ acontecer, considerando todas as possíveis hipóteses.

> "**Partições** são divisões do espaço amostral".

## Intuição do Teorema de Bayes

O Teorema de Bayes nos ajuda a ajustar nossas expectativas sobre a ocorrência de um evento à luz de novas informações. De maneira simplificada, ele combina nosso conhecimento prévio (priori) com a nova evidência que coletamos (verossimilhança), para atualizar nossa estimativa da probabilidade de um evento (posteriori).

> Sempre que se tem uma condicional, o espaço amostral se restringe ao dado que se tem.

## Aplicações do Teorema de Bayes

O Teorema de Bayes é aplicado em diversas áreas, incluindo:

1. **Medicina e Diagnóstico**: Avaliação de testes clínicos para diagnosticar doenças, calculando a probabilidade de um paciente estar doente com base no resultado do teste e nas probabilidades prévias.

2. **Machine Learning e Inteligência Artificial**: Algoritmos de aprendizado supervisionado, como a classificação Bayesiana, são baseados em princípios bayesianos para fazer previsões.

3. **Economia e Finanças**: Utilizado para modelar incertezas em mercados financeiros e para prever eventos econômicos.

4. **Processamento de Linguagem Natural**: Na construção de filtros de spam e sistemas de recomendação.

## Exemplo Prático

Suponha que você queira calcular a probabilidade de uma pessoa ter uma doença $D$ dado que o teste para essa doença deu positivo ($T^+$). Suponha que:

- A probabilidade de uma pessoa ter a doença é $P(D) = 0,01$ (1%).
- A probabilidade de o teste ser positivo se a pessoa tem a doença é $P(T^+|D) = 0,99$ (99%).
- A probabilidade de o teste ser positivo se a pessoa **não** tem a doença é $P(T^+|D^c) = 0,05$ (5%).

A probabilidade de testar positivo, $P(T^+)$, pode ser calculada usando a **lei das probabilidades totais**:

$$
P(T^+) = P(T^+|D) \cdot P(D) + P(T^+|D^c) \cdot P(D^c)
$$
$$
P(T^+) = (0,99 \cdot 0,01) + (0,05 \cdot 0,99) = 0,0594
$$

Agora, aplicando o Teorema de Bayes para encontrar a probabilidade de a pessoa ter a doença dado que o teste foi positivo:

$$
P(D|T^+) = \frac{P(T^+|D) \cdot P(D)}{P(T^+)} = \frac{0,99 \cdot 0,01}{0,0594} \approx 0,1667
$$

Portanto, mesmo com um teste positivo, a probabilidade de a pessoa realmente ter a doença é de cerca de 16,67%.

## Conclusão

O Teorema de Bayes é uma ferramenta poderosa para atualizar probabilidades baseadas em novas evidências. Ele é fundamental em situações onde as condições são incertas e precisamos tomar decisões com base em informações incompletas.

### Teorema de Bayes com Partições

Quando temos partições no espaço amostral, o Teorema de Bayes pode ser adaptado para considerar essas partições. Uma **partição** é um conjunto de eventos mutuamente exclusivos e coletivamente exaustivos que cobrem todo o espaço amostral, ou seja, cada resultado possível pertence a exatamente um desses eventos.

## Fórmula do Teorema de Bayes com Partições

O Teorema de Bayes, levando em conta uma partição formada pelos eventos $A_1, A_2, \dots, A_n$ e um evento $B$, é expresso como:

$$
P(A_i|B) = \frac{P(B|A_i) \cdot P(A_i)}{\sum_{j=1}^{n} P(B|A_j) \cdot P(A_j)}
$$

### Explicação
- **Partição**: $A_1, A_2, \dots, A_n$ são eventos que formam uma partição do espaço amostral, ou seja, eles são mutuamente exclusivos ($A_i \cap A_j = \emptyset$ para $i \neq j$) e coletivamente exaustivos ($\bigcup_{i=1}^{n} A_i = \Omega$, onde $\Omega$ é o espaço amostral).
- **Probabilidade posterior**: $P(A_i|B)$ é a probabilidade de $A_i$ dado que o evento $B$ ocorreu.
- **Probabilidade verossímil**: $P(B|A_i)$ é a probabilidade de $B$ ocorrer dado que $A_i$ é verdadeiro.
- **Probabilidade anterior**: $P(A_i)$ é a probabilidade de $A_i$ antes de qualquer informação sobre $B$.
- **Probabilidade marginal**: A soma $\sum_{j=1}^{n} P(B|A_j) \cdot P(A_j)$ é a probabilidade total de $B$, considerando todas as possíveis partições.

## Aplicação com Partições

Quando trabalhamos com partições no contexto do Teorema de Bayes, usamos a **lei das probabilidades totais** para calcular a probabilidade de um evento $B$ como uma combinação das probabilidades condicionais dos eventos que formam a partição.

### Exemplo Prático com Partições

Vamos considerar um exemplo onde uma empresa tem três fábricas ($A_1$, $A_2$, $A_3$) que produzem um certo produto. As probabilidades de cada fábrica produzir o produto são:
- $P(A_1) = 0,5$
- $P(A_2) = 0,3$
- $P(A_3) = 0,2$

Sabemos que a probabilidade de um produto ser defeituoso, dado que veio de cada fábrica, é:
- $P(B|A_1) = 0,02$
- $P(B|A_2) = 0,03$
- $P(B|A_3) = 0,04$

Queremos calcular a probabilidade de que um produto defeituoso tenha sido fabricado pela fábrica $A_2$.

Primeiro, calculamos a probabilidade total de que um produto seja defeituoso usando a lei das probabilidades totais:

$$
P(B) = P(B|A_1) \cdot P(A_1) + P(B|A_2) \cdot P(A_2) + P(B|A_3) \cdot P(A_3)
$$
$$
P(B) = (0,02 \cdot 0,5) + (0,03 \cdot 0,3) + (0,04 \cdot 0,2) = 0,029
$$

Agora, aplicando o Teorema de Bayes para calcular $P(A_2|B)$, a probabilidade de que o produto defeituoso tenha sido fabricado pela fábrica $A_2$:

$$
P(A_2|B) = \frac{P(B|A_2) \cdot P(A_2)}{P(B)} = \frac{0,03 \cdot 0,3}{0,029} \approx 0,3103
$$

Portanto, a probabilidade de que um produto defeituoso tenha sido fabricado pela fábrica $A_2$ é de aproximadamente 31,03%.

## Conclusão

Ao considerar partições no espaço amostral, o Teorema de Bayes permite distribuir a probabilidade total de um evento com base nas probabilidades de cada parte da partição. Isso é especialmente útil em cenários onde os dados são divididos em várias categorias, permitindo uma análise mais precisa e detalhada das probabilidades condicionais.

$ 0,10 \cdot (0,16 + 0,10 - \Pr[\text{Lyme} \cap \text{HGE}]) = \Pr[\text{Lyme} \cap \text{HGE}] $

$$
0,10 \cdot (0,26 - \Pr[\text{Lyme} \cap \text{HGE}]) = \Pr[\text{Lyme} \cap \text{HGE}]

$$
0,026 - 0,10 \cdot \Pr[\text{Lyme} \cap \text{HGE}] = \Pr[\text{Lyme} \cap \text{HGE}]

0,026 = \Pr[\text{Lyme} \cap \text{HGE}] + 0,10 \cdot \Pr[\text{Lyme} \cap \text{HGE}]

0,026 = 1,10 \cdot \Pr[\text{Lyme} \cap \text{HGE}]

     - $ \Pr[A] = 0,16 $ (16% dos carrapatos têm Lyme)
     - $ \Pr[B] = 0,10 $ (10% dos carrapatos têm HGE)
     - 10% dos carrapatos que têm Lyme ou HGE têm ambas as doenças.

2. **Usando a fórmula da união de dois eventos**:
   A fórmula para a probabilidade da união de dois eventos é dada por:
   $$
   \Pr[A \cup B] = \Pr[A] + \Pr[B] - \Pr[A \cap B]
   $$
   Onde $ \Pr[A \cap B] $ é a probabilidade de um carrapato ser portador de ambas as doenças.

3. **Identificando a interseção**:
   O problema afirma que 10% dos carrapatos que têm Lyme ou HGE têm ambas as doenças. Assim, podemos expressar isso como:
   $$
   \Pr[A \cap B] = 0,10 \cdot \Pr[A \cup B]
   $$

4. **Substituindo na fórmula da união**:
   Podemos substituir a expressão de $ \Pr[A \cap B] $ na fórmula da união:
   $$
   0,10 \cdot \Pr[A \cup B] = \Pr[A \cap B]
   $$
   Agora, substituímos $ \Pr[A \cup B] $ pela fórmula:
   $$
   0,10 \cdot \left( \Pr[A] + \Pr[B] - \Pr[A \cap B] \right) = \Pr[A \cap B]
   $$

5. **Substituindo os valores conhecidos**:
   Substituindo $ \Pr[A] $ e $ \Pr[B] $:
   $$
   0,10 \cdot (0,16 + 0,10 - \Pr[A \cap B]) = \Pr[A \cap B]
   $$
   Ou seja:
   $$
   0,10 \cdot (0,26 - \Pr[A \cap B]) = \Pr[A \cap B]
   $$

6. **Resolvendo a equação**:
   Expandindo e reorganizando a equação:
   $$
   0,026 - 0,10 \cdot \Pr[A \cap B] = \Pr[A \cap B]
   $$
   $$
   0,026 = \Pr[A \cap B] + 0,10 \cdot \Pr[A \cap B]
   $$
   $$
   0,026 = 1,10 \cdot \Pr[A \cap B]
   $$
   Agora isolamos $ \Pr[A \cap B] $:
   $$
   \Pr[A \cap B] = \frac{0,026}{1,10} = 0,0236
   $$

\Pr[\text{Lyme} \cap \text{HGE}] = 0,0236

2. Suponha que para a população em geral, **1** em cada **5000** pessoas carrega o vírus HIV e que 
existe um teste para a presença do HIV cujo os possíveis resultados são `positivo (+)` e `negativo 
(−)`. **Suponha que o teste acerte a resposta em 99% dos casos**, isto é, para uma pessoa que tem 
HIV ou não, o teste dá o resultado correto 99% das vezes.
- a) Qual é a probabilidade 𝑃𝑟[−|𝐻𝐼𝑉] de que o teste dê negativo dado que a pessoa possui o vírus HIV?
- b) Qual a probabilidade 𝑃𝑟[𝐻𝐼𝑉|+] de que uma pessoa escolhida aleatoriamente tenha o vírus HIV dado que o teste deu positivo?

>###### Resolução do Problema de Probabilidade com HIV

### Dados do problema:

- Prevalência do HIV: $ P(HIV) = \frac{1}{5000} = 0,0002 $ (0,02% da população tem HIV).
- Probabilidade de testar positivo se a pessoa tem HIV (verdadeiro positivo): $ P(+ | HIV) = 0,99 $.
- Probabilidade de testar negativo se a pessoa não tem HIV (verdadeiro negativo): $ P(- | \neg HIV) = 0,99 $.

### Parte 1: Probabilidade de testar negativo dado que a pessoa tem HIV

A probabilidade de um teste dar negativo dado que a pessoa tem HIV é a complementação da probabilidade de um teste dar positivo para quem tem HIV:
$$
P(- | HIV) = 1 - P(+ | HIV) = 1 - 0,99 = 0,01
$$

**Portanto, a probabilidade de que o teste dê negativo dado que a pessoa possui o vírus HIV é:**
$$
P(- | HIV) = 0,01
$$

---

### Parte 2: Probabilidade de ter HIV dado que o teste deu positivo

Para calcular essa probabilidade, usaremos o Teorema de Bayes:
$$
P(HIV | +) = \frac{P(+ | HIV) \cdot P(HIV)}{P(+)}
$$

Onde $ P(+) $ (probabilidade total de testar positivo) pode ser calculada considerando dois casos:
1. A pessoa tem HIV e o teste acerta (positivo).
2. A pessoa não tem HIV e o teste erra (falso positivo).

**Cálculo de $ P(+) $:**
$$
P(+) = P(+ | HIV) \cdot P(HIV) + P(+ | \neg HIV) \cdot P(\neg HIV)
$$
Sabemos que:
- $ P(\neg HIV) = 1 - P(HIV) = 1 - 0,0002 = 0,9998 $
- A probabilidade de testar positivo para quem não tem HIV (falso positivo) é $ P(+ | \neg HIV) = 1 - P(- | \neg HIV) = 1 - 0,99 = 0,01 $.

Substituindo os valores:
$$
P(+) = (0,99 \cdot 0,0002) + (0,01 \cdot 0,9998)
$$
$$
P(+) = 0,000198 + 0,009998 = 0,010196
$$

Agora, podemos aplicar o Teorema de Bayes:
$$
P(HIV | +) = \frac{P(+ | HIV) \cdot P(HIV)}{P(+)} = \frac{0,99 \cdot 0,0002}{0,010196}
$$
Calculando:
$$
P(HIV | +) = \frac{0,000198}{0,010196} \approx 0,0194
$$

**Portanto, a probabilidade de que uma pessoa escolhida aleatoriamente tenha o vírus HIV dado que o teste deu positivo é aproximadamente:**
$$
P(HIV | +) \approx 0,0194 \text{ ou } 1,94\%
$$

### Resumo dos Resultados:
- $ P(- | HIV) = 0,01 $ (1%)
- $ P(HIV | +) \approx 0,0194 $ (1,94%)


```julia
# Dados do problema
P_HIV = 1 / 5000       # Prevalência do HIV
P_pos_HIV = 0.99       # Probabilidade de teste positivo dado que a pessoa tem HIV
P_neg_not_HIV = 0.99   # Probabilidade de teste negativo dado que a pessoa não tem HIV

# Cálculo da probabilidade de teste negativo dado que a pessoa tem HIV
P_neg_HIV = 1 - P_pos_HIV
println("Probabilidade de que o teste dê negativo dado que a pessoa possui HIV: P(- | HIV) = $P_neg_HIV")

# Cálculo da probabilidade de não ter HIV
P_not_HIV = 1 - P_HIV

# Cálculo da probabilidade de teste positivo dado que a pessoa não tem HIV
P_pos_not_HIV = 1 - P_neg_not_HIV

# Cálculo da probabilidade total de testar positivo
P_pos = (P_pos_HIV * P_HIV) + (P_pos_not_HIV * P_not_HIV)

# Cálculo da probabilidade de ter HIV dado que o teste deu positivo (Teorema de Bayes)
P_HIV_pos = (P_pos_HIV * P_HIV) / P_pos
println("Probabilidade de que uma pessoa escolhida aleatoriamente tenha o vírus HIV dado que o teste deu positivo: P(HIV | +) = $P_HIV_pos")
```

Para calcular o cenário inverso, onde precisamos encontrar a probabilidade de que uma pessoa não tenha HIV dado que o teste deu positivo ($ P(\neg HIV | +) $), podemos usar o Teorema de Bayes da seguinte forma:

### Passos para o Cálculo:

1. **Definir as probabilidades necessárias**:
$$ P(\neg HIV) = 1 - P(HIV) $$
$$ P(+ | \neg HIV) = 1 - P(- | \neg HIV) $$

2. **Calcular $ P(+) $**:
$$ P(+) = P(+ | HIV) \cdot P(HIV) + P(+ | \neg HIV) \cdot P(\neg HIV) $$

3. **Aplicar o Teorema de Bayes**:
$$ P(\neg HIV | +) = \frac{P(+ | \neg HIV) \cdot P(\neg HIV)}{P(+)} $$

```julia
# Dados do problema
P_HIV = 1 / 5000       # Prevalência do HIV
P_pos_HIV = 0.99       # Probabilidade de teste positivo dado que a pessoa tem HIV
P_neg_not_HIV = 0.99   # Probabilidade de teste negativo dado que a pessoa não tem HIV

# Probabilidades necessárias
P_not_HIV = 1 - P_HIV                      # Probabilidade de não ter HIV
P_pos_not_HIV = 1 - P_neg_not_HIV          # Probabilidade de teste positivo dado que a pessoa não tem HIV

# Cálculo da probabilidade total de testar positivo
P_pos = (P_pos_HIV * P_HIV) + (P_pos_not_HIV * P_not_HIV)

# Cálculo da probabilidade de não ter HIV dado que o teste deu positivo (Teorema de Bayes)
P_not_HIV_pos = (P_pos_not_HIV * P_not_HIV) / P_pos

println("Probabilidade de que uma pessoa escolhida aleatoriamente não tenha o vírus HIV dado que o teste deu positivo: P(¬HIV | +) = $P_not_HIV_pos")
```

3. Suponha que em uma fábrica, **2** em cada **1000** produtos `são defeituosos`. A fábrica realiza um teste de qualidade que pode resultar em `defeituoso (D)` ou em `perfeitas condições (E)`. Suponha que o teste acerte a resposta em **95% dos casos**, isto é, para um produto que é defeituoso ou não, o teste dá o resultado correto 95% das vezes.

- a) Qual é a probabilidade $P(D∣E)$ de que o teste dê resultado `defeituoso` dado que o produto está em `perfeitas condições`?

- b) Qual a probabilidade $P(D∣D)$ de que um produto escolhido aleatoriamente seja `defeituoso` dado que o teste deu resultado `defeituoso?

> ###### Resolução

Dado:
- A probabilidade de um produto ser defeituoso é: $P(D) = \frac{2}{1000} = 0,002$.
- A probabilidade de um produto estar em perfeitas condições é: $P(E) = 1 - P(D) = 0,998$.
- O teste de qualidade é correto em 95% dos casos, ou seja:
  - Se o produto é defeituoso, a probabilidade de o teste indicar defeituoso é: $P(T_d \mid D) = 0,95$.
  - Se o produto é perfeito, a probabilidade de o teste indicar perfeito é: $P(T_e \mid E) = 0,95$.

Portanto:
- A probabilidade de o teste indicar defeituoso para um produto perfeito é: $P(T_d \mid E) = 1 - P(T_e \mid E) = 0,05$.
- A probabilidade de o teste indicar perfeito para um produto defeituoso é: $P(T_e \mid D) = 1 - P(T_d \mid D) = 0,05$.

#### Questão (a)

**Qual é a probabilidade $P(D \mid E)$ de que o teste dê resultado "defeituoso" dado que o produto está em perfeitas condições?**

A probabilidade $P(D \mid E)$ não se aplica diretamente, mas podemos calcular a probabilidade de que o teste indique defeituoso quando o produto é perfeito usando a regra dada anteriormente:

$$
P(T_d \mid E) = 0,05
$$

Portanto, a resposta para a questão (a) é:
$$
P(D \mid E) = 0,05
$$

#### Questão (b)

**Qual a probabilidade $P(D \mid D)$ de que um produto escolhido aleatoriamente seja defeituoso dado que o teste deu resultado "defeituoso"?**

Aplicamos o Teorema de Bayes para calcular essa probabilidade:

$$
P(D \mid T_d) = \frac{P(T_d \mid D) \cdot P(D)}{P(T_d)}
$$

$$
P(D \mid T_d) = \frac{0,95 \cdot 0,002}{0,0518}

P(D \mid T_d) = \frac{0,0019}{0,0518} \approx 0,0367

5. Uma fábrica de blocos estruturais possui três setores, denominados $A$, $B$ e $C$, que respondem, respectivamente, por $45$%, $35$% e $20$% da produção. Sabendo-se que os setores $A$, $B$ e $C$ produzem, respectivamente, $2$%, $4$% e $5$% de seus blocos com defeitos.

b) **Sendo encontrado um bloco com defeito, qual a probabilidade de ele ter vindo do Bloco C?**

> a) Para resolver essa questão, devemos usar o **Teorema da Probabilidade Total**. Esse teorema permite calcular a probabilidade de um evento (neste caso, um bloco defeituoso) considerando várias fontes que podem causar esse evento (os setores A, B e C).

### Definindo as variáveis:

- $ P(A) = 0,45 $ (probabilidade de o bloco vir do setor A)
- $ P(B) = 0,35 $ (probabilidade de o bloco vir do setor B)
- $ P(C) = 0,20 $ (probabilidade de o bloco vir do setor C)

- $ P(D|A) = 0,02 $ (probabilidade de um bloco do setor A ser defeituoso)
- $ P(D|B) = 0,04 $ (probabilidade de um bloco do setor B ser defeituoso)
- $ P(D|C) = 0,05 $ (probabilidade de um bloco do setor C ser defeituoso)

### Probabilidade total de um bloco ser defeituoso ($P(D)$):

$$
P(D) = P(A) \cdot P(D|A) + P(B) \cdot P(D|B) + P(C) \cdot P(D|C)
$$

Substituindo os valores:

$$
P(D) = (0,45 \times 0,02) + (0,35 \times 0,04) + (0,20 \times 0,05)
$$

$$
P(D) = 0,009 + 0,014 + 0,01
$$

$$
P(D) = 0,033
$$

### Resultado:

A probabilidade de um bloco qualquer, produzido na fábrica, ser defeituoso é **0,033**, ou seja, **3,3%**.


> b) Essa questão pode ser resolvida utilizando o **Teorema de Bayes**, que calcula a probabilidade condicional de um evento dado outro evento já ocorrido.

### O que queremos calcular:
A probabilidade de o bloco defeituoso ter vindo do setor C, dado que o bloco encontrado é defeituoso. Isso é representado por $P(C|D)$.

### Teorema de Bayes:
$$
P(C|D) = \frac{P(C) \cdot P(D|C)}{P(D)}
$$
Onde:
- $P(C|D)$ é a probabilidade de o bloco vir do setor C, dado que ele é defeituoso.
- $P(C) = 0,20$ é a probabilidade de o bloco ter vindo do setor C.
- $P(D|C) = 0,05$ é a probabilidade de um bloco do setor C ser defeituoso.
- $P(D) = 0,033$ é a probabilidade total de um bloco ser defeituoso (calculada anteriormente).

### Substituindo os valores:

$$
P(C|D) = \frac{0,20 \times 0,05}{0,033}
$$

$$
P(C|D) = \frac{0,01}{0,033}
$$

$$
P(C|D) \approx 0,303
$$

### Resultado:

A probabilidade de o bloco defeituoso ter vindo do setor C é aproximadamente **30,3\%**.

### **Q1 – Cálculo da probabilidade de uma pessoa receber mais de $83,00 por hora**

#### **Passo 1: Determinar os parâmetros da distribuição normal**

##### Cálculo da Média ($\mu$):

A média ponderada é dada por:

$$
\mu = \frac{\sum (x_i \cdot f_i)}{n}
$$

Onde:
- $x_i$ são os valores centrais de cada intervalo,
- $f_i$ é o número de pessoas em cada intervalo,
- $n$ é o total de pessoas, $n = 100$.

Calculando:

$$
\mu = \frac{(55 \cdot 15) + (65 \cdot 20) + (75 \cdot 30) + (85 \cdot 20) + (95 \cdot 15)}{100}
$$

$$
\mu = \frac{(825) + (1300) + (2250) + (1700) + (1425)}{100} = \frac{7500}{100} = 75
$$

Portanto, a média ($\mu$) dos salários é **75**.

##### Cálculo do Desvio Padrão ($\sigma$):

O desvio padrão é dado por:

$$
\sigma = \sqrt{\frac{\sum f_i (x_i - \mu)^2}{n}}
$$

Calculando:

$$
\sigma = \sqrt{\frac{15(55 - 75)^2 + 20(65 - 75)^2 + 30(75 - 75)^2 + 20(85 - 75)^2 + 15(95 - 75)^2}{100}}
$$

$$
\sigma = \sqrt{\frac{15(400) + 20(100) + 30(0) + 20(100) + 15(400)}{100}}
$$

$$
\sigma = \sqrt{\frac{6000 + 2000 + 0 + 2000 + 6000}{100}} = \sqrt{\frac{16000}{100}} = \sqrt{160} \approx 12,65
$$

Portanto, o desvio padrão ($\sigma$) é aproximadamente **12,65**.

#### **Passo 2: Calcular o valor de Z**

Agora que temos a média e o desvio padrão, podemos calcular a probabilidade usando a fórmula do valor padronizado **Z**:

$$
Z = \frac{X - \mu}{\sigma}
$$

Onde:
- $X = 83$ (o valor do salário que estamos analisando),
- $\mu = 75$ (a média),
- $\sigma = 12,65$ (o desvio padrão).

Calculando $Z$:

$$
Z = \frac{83 - 75}{12,65} = \frac{8}{12,65} \approx 0,632
$$

#### **Passo 3: Consultar a tabela da distribuição normal**

Com $Z = 0,632$, precisamos consultar a **tabela da distribuição normal padrão** ou usar uma calculadora para encontrar a área à esquerda desse valor.

Pela tabela, a probabilidade associada a $Z = 0,632$ é aproximadamente **0,736**. Isso significa que a probabilidade de uma pessoa receber **menos de $83,00** é 0,736 ou 73,6%.

#### **Passo 4: Calcular a probabilidade de receber mais de $83,00**

A probabilidade de uma pessoa receber mais de $83,00 por hora é o complemento da probabilidade de receber menos:

$$
P(X > 83) = 1 - P(X \leq 83) = 1 - 0,736 = 0,264
$$

Portanto, a **probabilidade de uma pessoa receber mais de $83,00** por hora é **0,264** ou **26,4%**.

2. Em um restaurante, três cozinheiros **A**, **B** e **C** preparam um tipo especial de bolo, com probabilidades respectivas de $0,02$, $0,03$ e $0,05$ de que a massa do bolo não cresça. Sabe-se que **A** prepara 50% dos bolos, **B** 30% e **C** 20%. 

Se uma massa de bolo não cresceu, qual a probabilidade de que tenha sido preparada pelo cozinheiro **A**?

### **Q2 – Probabilidade de que o bolo foi preparado pelo cozinheiro A**

#### **Passo 1: Definir os eventos e probabilidades**

Temos três cozinheiros $A$, $B$ e $C$ que preparam bolos, e as probabilidades de que um bolo não cresça variam entre eles:

- $P(A) = 0,50$ (50% dos bolos são feitos por $A$),
- $P(B) = 0,30$ (30% dos bolos são feitos por $B$),
- $P(C) = 0,20$ (20% dos bolos são feitos por $C$).

As probabilidades de que o bolo não cresça para cada cozinheiro são:

- $P(\text{não crescer} | A) = 0,02$,
- $P(\text{não crescer} | B) = 0,03$,
- $P(\text{não crescer} | C) = 0,05$.

#### **Passo 2: Aplicar o Teorema de Bayes**

Queremos calcular a probabilidade de que o bolo tenha sido preparado pelo cozinheiro $A$, dado que ele **não cresceu**. Para isso, usamos o **Teorema de Bayes**:

$$
P(A | \text{não crescer}) = \frac{P(\text{não crescer} | A) \cdot P(A)}{P(\text{não crescer})}
$$

Onde $P(\text{não crescer})$ é a probabilidade total de um bolo não crescer, dada por:

$$
P(\text{não crescer}) = P(\text{não crescer} | A) \cdot P(A) + P(\text{não crescer} | B) \cdot P(B) + P(\text{não crescer} | C) \cdot P(C)
$$

Calculando:

$$
P(\text{não crescer}) = (0,02 \cdot 0,50) + (0,03 \cdot 0,30) + (0,05 \cdot 0,20)
$$

$$
P(\text{não crescer}) = 0,01 + 0,009 + 0,01 = 0,029
$$

#### **Passo 3: Calcular a probabilidade de $A$ dado que o bolo não cresceu**

Agora, aplicando o Teorema de Bayes:

$$
P(A | \text{não crescer}) = \frac{0,02 \cdot 0,50}{0,029} = \frac{0,01}{0,029} \approx 0,3448
$$

Portanto, a **probabilidade de que o bolo tenha sido preparado pelo cozinheiro A**, dado que ele não cresceu, é **aproximadamente 34,48%**.
